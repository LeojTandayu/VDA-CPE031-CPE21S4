# **Types of data analysis**



**Descriptive Analysis**
- is the most common and primary forms of data analysis. It is used to find the "what is happening?" in business. Usually, we take to help of descriptive data analysis to track the Key Performance Indicator (KPI), sales profit/loss.


**Diagnostic Data Analysis**
- Diagnostic data analysis helps find out the "Why did something happen?" Once we get the report of what happened from desciptive analysis, diagnostic data analysis helps us understand which factors caused something to happen.

**Prescriptive Data Analysis**
- Prescriptive analysis uses the outcomes of all the data analysis. It will help fid the "what action should be taken?" to counter a problem or predict a problem. So it prescribes the action/s as a solution to counter the specific situation. It needs advance machine learning and real-time AI.
