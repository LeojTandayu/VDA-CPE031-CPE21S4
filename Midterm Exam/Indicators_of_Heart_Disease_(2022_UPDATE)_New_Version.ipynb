1. What is the statistics of patients who is smoking, alcohol drinking all according to their age which is shown by a piechart?

```python
import matplotlib.pyplot as plt
import pandas as pd

patients = pd.read_csv('heart_2022_no_nans.csv')

smoking_stats = patients.groupby(['AgeCategory', 'SmokerStatus'])['SmokerStatus'].count().reset_index(name='count')
plt.figure(figsize=(16, 6))
plt.bar(smoking_stats['AgeCategory'] + ' - ' + smoking_stats['SmokerStatus'], smoking_stats['count'])
plt.title('Smoking Statistics by Age Category')
plt.xlabel('Age Category and Smoking Status')
plt.ylabel('Number of Patients')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

alcohol_stats = patients.groupby(['AgeCategory', 'AlcoholDrinkers'])['AlcoholDrinkers'].count().reset_index(name='count')
plt.figure(figsize=(16, 6))
plt.bar(alcohol_stats['AgeCategory'] + ' - ' + alcohol_stats['AlcoholDrinkers'], alcohol_stats['count'])
plt.title('Alcohol Drinking Statistics by Age Category')
plt.xlabel('Age Category and Alcohol Drinking Status')
plt.ylabel('Number of Patients')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

There are 2 separate graphs showing the number of patients based on their age who are drinking alcohol and smoking. The ages range from 18-24, 25-29, 30-34, up to 80 or older. Based on the graph, it shows that majority of patients in all the age range has never experienced smoking. On the alcohol drinking, as the age goes higher up to the 65-69 years old patients, the numbers of alcohol drinkers increases. And it decreases as the goes above 70.

2. What are the statistics of the number of patients who are at high risk based on their gender? Using bar graph.

```python
import matplotlib.pyplot as plt
import pandas as pd

patients = pd.read_csv('heart_2022_no_nans.csv')

high_risk_patients = patients[patients['HighRiskLastYear'] == 'Yes']

gender_stats = high_risk_patients.groupby('Sex')['Sex'].count().reset_index(name='count')

plt.figure(figsize=(6, 6))
plt.bar(gender_stats['Sex'], gender_stats['count'])
plt.title('High-Risk Patients by Gender (HighRiskLastYear)')
plt.xlabel('Gender')
plt.ylabel('Number of Patients')
plt.show()
```

According to the graph that is shown, majority of the patients who are at risk based on their genders are male which is above 6000 patients, followed by female who are above 4000 patients
