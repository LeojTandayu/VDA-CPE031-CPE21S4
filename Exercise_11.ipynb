```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tabulate import tabulate

x = np.array([0.50,1.00,1.30,1.60,2.00,2.20,2.50,3.10,3.90,4.40])
y = np.array([1.90,1.50,1.20,1.00,0.80,0.78,0.65,0.46,0.30,0.23])


df = pd.DataFrame({'x': x, 'y': y})
print(tabulate(df , headers = "keys" , tablefmt='psql'))

df.to_csv('student_data.csv', index=False)

#Calculate the slope and y-interception of the regression line
slope, intercept = np.polyfit(x, y, 1)

#Calculate the predicted values of y using the regression line equation
y_pred = slope * x + intercept

#Plot the data and the regression line
plt.scatter(x, y)
plt.plot(x, slope * x + intercept, color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

df.to_csv('Sample Problem.csv', index=False)
```

The given set of x and y data formed a negative correlation. As shown in the plot, all of the points are close to the regression line. It shows that as the x value increases, the y value decreases which can be seen in the table and the plot.
