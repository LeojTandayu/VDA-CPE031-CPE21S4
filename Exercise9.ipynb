- show the legend at the lower right location
- x label name = years
- y label name = sales

- add a circle marker

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Technological-Products-Sample-Data - Tech Products.csv')
plt.figure(figsize=(15, 5))

plt.title('Price of Technological Products Over Time')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.plot(df['Date of Release'], df['Price (USD)'], linestyle = 'dotted', color = 'r', label = 'Total Sales', marker = 'o', linewidth = 3)
plt.legend(loc='lower right')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

c. Read all the products sales data and show it using a multiline plot. Label them.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Technological-Products-Sample-Data - Tech Products.csv')

df['Date of Release'] = pd.to_datetime(df['Date of Release'])
df['Release Month'] = df['Date of Release'].dt.to_period('M')

plt.figure(figsize=(15, 5))
plt.title('Sales Data for All Products')
plt.xlabel('Release Month')
plt.ylabel('Total Sales (USD)')

for product in df['Model'].unique():
    product_data = df[df['Model'] == product].groupby('Release Month')['Price (USD)'].sum().reset_index()
    plt.plot(product_data['Release Month'].astype(str), product_data['Price (USD)'], label=product)

plt.xticks(rotation=45, ha='right')
plt.legend(loc='best', fontsize='small')
plt.tight_layout()
plt.show()
```

f. Create a pie chart that shows each brand country of origin. Label them

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Technological-Products-Sample-Data - Tech Products.csv')

country_counts = df['Country of Origin'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(country_counts, labels=country_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Country of Origin for Brands')
plt.axis('equal')

plt.show()
```
