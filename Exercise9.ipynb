b. Get the total sales and show line plot with the followiung style properties.

- line style dotted and line color should be red

- show the legend at the lower right location
- x label name = years
- y label name = sales

- add a circle marker
- line marker color as read
- line width should be 3

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Technological-Products-Sample-Data - Tech Products.csv')
plt.figure(figsize=(15, 5))

plt.title('Price of Technological Products Over Time')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.plot(df['Date of Release'], df['Price (USD)'], linestyle = 'dotted', color = 'r', label = 'Total Sales', marker = 'o', linewidth = 3)
plt.legend(loc='lower right')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

c. Read all the products sales data and show it using a multiline plot. Label them.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Technological-Products-Sample-Data - Tech Products.csv')

plt.figure(figsize=(15, 5))
plt.title('Sales Data for All Products')
plt.xlabel('Date of Release')
plt.ylabel('Sales')

for product in df['Model'].unique():
    product_data = df[df['Model'] == product]
    plt.plot(product_data['Date of Release'], product_data['Price (USD)'], label=product)

plt.xticks(rotation=45)
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```
